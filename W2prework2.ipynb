問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
import numpy as np
from numpy.random import *

np.random.seed(0)
mu = [-3,0]
sigma = [[1.0, 0.8],[0.8, 1.0]]

sample_ndarray = np.random.multivariate_normal(mu, sigma,500)

print(sample_ndarray.shape)
```

【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(sample_ndarray[:,0],sample_ndarray[:,1])
plt.show()
```

【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(2)
axes[0].hist(sample_ndarray[:,0],range=(-5,3), bins=50)
axes[0].set_title("x1 histogram")
axes[0].set_xlabel("x1")
axes[0].set_ylabel("frequency")
axes[1].hist(sample_ndarray[:,1],range=(-5,3), bins=50)
axes[1].set_title("x2 histogram")
axes[1].set_xlabel("x2")
axes[1].set_ylabel("frequency")
fig.tight_layout()
plt.show()
```

【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
np.random.seed(0)
mu_1 = [-3,0]
mu_2 = [0,-3]
sigma = [[1.0, 0.8],[0.8, 1.0]]

sample_ndarray_1 = np.random.multivariate_normal(mu_1, sigma, 500)
sample_ndarray_2 = np.random.multivariate_normal(mu_2, sigma, 500)
print(sample_ndarray_1.shape)
print(sample_ndarray_2.shape)

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(sample_ndarray_1[:,0],sample_ndarray_1[:,1], label="0")
ax.scatter(sample_ndarray_2[:,0],sample_ndarray_2[:,1], label="1")
ax.set_title("scatter")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.legend()

plt.show()
```

【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを結合 し、
(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
sample_ndarray_c = np.concatenate([sample_ndarray_1, sample_ndarray_2])
print("ndarray1:",sample_ndarray_1.shape)
print("ndarray2:",sample_ndarray_2.shape)
print("結合したndarray",sample_ndarray_c.shape)
```
